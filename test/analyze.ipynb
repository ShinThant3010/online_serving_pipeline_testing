```python
import pandas as pd
```

```python
feeds_5k = pd.read_json("data/feeds_5k.jsonl", lines=True)
print(len(feeds_5k))
feeds_5k.head()
```

```python
feeds_20k = pd.read_json("data/feeds_20k.jsonl", lines=True)
print(len(feeds_20k))
feeds_20k.head()
```

```python

# Find common feed_ids
common_feed_ids = set(feeds_5k["feed_id"]).intersection(feeds_20k["feed_id"])

print(f"Common feed_ids count: {len(common_feed_ids)}")

# Optional: view as sorted list
common_feed_ids_list = sorted(common_feed_ids)
print(common_feed_ids_list[:20])  # first 20
```

```python
common_feeds = feeds_5k[feeds_5k["feed_id"].isin(common_feed_ids)]
print(len(common_feeds))
common_feeds.head()
```

```python
common_feeds.to_json("data/common_feeds.jsonl", orient="records", lines=True)
```
